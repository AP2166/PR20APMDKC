### Do the jury and televoters agree on who should be the winner?

We briefly analyze this in the Performance analysis notebook, but a more detailed examination can be found in "Jury and televoting differences.ipynb". 

There's a very dramatic difference between the top 10 placements according to the jury compared to the opinion of the public. 
In 2019, it was an average of 6 places. In 2018, it was 8 places, in 2017 6 places and in 2016 it was 5 places.

When we comes to the placement of all performers, it was again, 6 places in 2019, 7 in 2018, 7 in 2017 and 8 in 2016.
We also compared the final placements (according to the combined points of the jury and the televoters) to the placements according to the jury and televoting, separately.

We found out that the televoting points matter a lot more. When it comes to the top 10, the televoters' placement is only off by an average of (less than) 3 places, whereas the jury misses by 4 places. When it comes to all the placements, the televoters are marginally better than the jury. 
<br>
We can account these results to the fact that the televoters are much more sure about the songs they like or don't like. The televoters are almost uninanimous in their decision about who should be the winner. Also, a lot of performers might end up getting almost no points by the televoters. The jury on the other hand is much more generous with their points and almost all performers get some points by at least some country's jury. This might be due to politics, but that's just a speculation. The standard deviation of the televoting points is always larger than the standard deviation of the jury's points. 

However, the support of the televoters isn't a guarantee for win, because for example in 2019, even though Norway got the most points by the televoters, the jury didn't like the song and it only gave them 40 points. Norway ended up sixth. The opposite also happened. The jury's favorite was North Macedonia, but the song wasn't a fan favorite. It ended up seventh.

When it comes to the diaspora effect question, we decided to examine 6 countries which have a large foreign population, namely Spain, Russia, Germany, the United Kingdom, Italy and Slovenia. We did this by first plotting a graph for each country where the televoting and jury points were divided. Then, we saw which countries got much more points from the public than from the jury. Then, by checking out official statistics for the foreign population in these countries we checked whether there's an overlap between the countries. And sure enough, there was an overlap in all the examples. Sure, the countries didn't overlap exactly, but more than 80 percent of the countries were the same in both lists. This explains the fact that Bulgaria seems to be the favourite country for both the United Kingdom and Spain.

### Prediction model

In the "models.ipynb" file, the code for building a prediction model can be found using the dataset of all performances in the finals from 2015 to 2019. First, we needed to see which attributes are the most important. Using a pearson correlation function, we found out that the (very subjective) song, performance and memorability ratings were the attributes most strongly correlated to the total points (which we wanted to predict).

```python
from IPython.display import Image
Image("correlation.jpg")
```

From this, we built three different regression models for each year: linear regression, lasso and ridge (25% test, 75% train splitting of the data).

The three models were more or less similarly successful in predicting the total points value, with a mean absolute error of 90. This may seem like a large error, but the maximum amount of points that can be won is bigger than 900, depending on the number of countries participating in a particular year, and this prediction model is accurate enough to categorize whether a song will or won't place high in the competition.

We also tried to create a prediction model which takes all the remaining attributes into account (choreography, props on stage etc.) but this model was a lot less successful so we decided not to include it.
